```python
from qiskit import QuantumCircuit
from qiskit import transpile
from qiskit import QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
```

```python
from qiskit.visualization import plot_histogram
from qiskit.visualization import circuit_drawer
import matplotlib.pyplot as plt
```

```python
import pylatexenc
```

```python
q = QuantumRegister(16, 'q')
c = ClassicalRegister(16, 'c')
circuit = QuantumCircuit(q, c)
# circuit=QuantumCircuit(16, 16)
```

```python
circuit
```

```python
# circuit.h(range(16))
circuit.h(q)
```

```python
circuit.measure(q, c)
# circuit.measure(range(16), range(16))
```

```python
circuit.draw()
```

```python
circuit_drawer(circuit, output='text')
```

```python
circuit_drawer(circuit, output='mpl')
# plt.show()
```

```python
pip install pylatexenc
```

```python
simulator=AerSimulator()
```

```python
transpiled_circuit=transpile(circuit, simulator)
# print(transpiled_circuit)
```

```python
print("Executing Job......")
job=simulator.run(transpiled_circuit, shots=10)
result=job.result()
print(result)
```

```python
counts = result.get_counts(transpiled_circuit)
print(counts)
```

```python
random_number_bin = list(counts.keys())[0]
random_number_bin
```

```python
random_number_dec = int(random_number_bin, 2)
random_number_dec
```

```python
plot_histogram(counts)
```
